### Research Question 1 (What is the Movie Gener that shown most?)

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.

#create new column with main genres#
movies['main_genres']= movies['genres'].str.split('|',expand=True)[0]

#pie chart show the count of each Movie Genre 
movies.groupby('main_genres')['id'].count().plot(figsize = (10,10), kind= 'Pie')
```

```python
movies.groupby('release_year').count()['id'].plot(kind='bar',figsize=(15,5))
```

### Research Question 2  (What is the average of votes for each Movie Genre over the years?)

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.

#split the dataset in to multiple datatsets as per Movie gener
Action = movies.query('main_genres == "Action"')
Animation= movies.query('main_genres == "Animation"')
Drama= movies.query('main_genres == "Drama"')
Science_Fiction = movies.query('main_genres == "Science Fiction"')
Comedy= movies.query('main_genres == "Comedy"')
Romance= movies.query('main_genres == "Romance"')
Thriller= movies.query('main_genres == "Thriller"')
Horror= movies.query('main_genres == "Horror"')
Crime= movies.query('main_genres == "Crime"')
War= movies.query('main_genres == "War"')

Action = Action.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)
Animation = Animation.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)
Drama = Drama.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)
Science_Fiction = Science_Fiction.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)
Comedy = Comedy.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)
Romance = Romance.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)
Thriller = Thriller.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)
Horror = Horror.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)
Crime = Crime.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)
War = War.groupby(['release_year', 'main_genres'], as_index=False)['vote_average','revenue'].aggregate(np.mean)


#chart show the vote average in each year for each Movie genre
Action.plot(x='release_year', y='vote_average', title = 'Action',marker='.')
Animation.plot(x='release_year', y='vote_average', title = 'Animation', marker='.')
Drama.plot(x='release_year', y='vote_average', title = 'Drama', marker='.')
Science_Fiction.plot(x='release_year', y='vote_average', title = 'Science Fiction', marker='.')
Comedy.plot(x='release_year', y='vote_average', title = 'Comedy', marker='.')
Romance.plot(x='release_year', y='vote_average', title = 'Romance', marker='.')
Thriller.plot(x='release_year', y='vote_average', title = 'Thriller', marker='.')
Horror.plot(x='release_year', y='vote_average', title = 'Horror', marker='.')
Crime.plot(x='release_year', y='vote_average', title = 'Crime', marker='.')
War.plot(x='release_year', y='vote_average', title = 'War', marker='.')
```

### Research Question 3  (What is the average of Revenue for each Movie Genre over the years?)


```python

#chart show the Revenue average in each year for each Movie genre
Action.plot(x='release_year', y='revenue', title = 'Action',marker='.')
Animation.plot(x='release_year', y='revenue', title = 'Animation', marker='.')
Drama.plot(x='release_year', y='revenue', title = 'Drama', marker='.')
Science_Fiction.plot(x='release_year', y='revenue', title = 'Science Fiction', marker='.')
Comedy.plot(x='release_year', y='revenue', title = 'Comedy', marker='.')
Romance.plot(x='release_year', y='revenue', title = 'Romance', marker='.')
Thriller.plot(x='release_year', y='revenue', title = 'Thriller', marker='.')
Horror.plot(x='release_year', y='revenue', title = 'Horror', marker='.')
Crime.plot(x='release_year', y='revenue', title = 'Crime', marker='.')
War.plot(x='release_year', y='revenue', title = 'War', marker='.')

```
